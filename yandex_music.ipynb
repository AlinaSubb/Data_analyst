Объявлю функцию *`correct_wrong_genre()`*, которая заменяет  неправильное название жанра в столбце *`genre`* на правильное и проверяет успешность выполнения замены.

```python
def correct_wrong_genre(df, wrong, right):
    df['genre'] = df['genre'].replace(wrong, right)
    count_wrong = df[df['genre'] == wrong]['genre'].count()
    print (f"{wrong} исправлен на {right}. Неправильных значений: {count_wrong}")
```

Исправлю различные варианты названия жанра хип-хоп в таблице (*`hip`*, *`hop`*, *`hip-hop`*) на *`hiphop`*.

```python
correct_wrong_genre(df, 'hip', 'hiphop')
correct_wrong_genre(df, 'hop', 'hiphop')
correct_wrong_genre(df, 'hip-hop', 'hiphop')
```

## Действительно ли музыку в разных городах слушают по-разному?

Была выдвинута гипотеза, что активность пользователей зависит от дня недели, причём в Москве и Петербурге это проявляется по-разному. Проверю это предположение по данным о трёх днях недели — понедельнике, среде и пятнице. Для каждого города устанавлю количество прослушанных  в эти дни композиций с известным жанром и сравнивню результаты.

Активность пользователей зависит от дня недели. Причём в Москве и Петербурге это проявляется по-разному.

```python
df.groupby('city')['genre'].count()
```

В Москве прослушиваний больше, чем в Питере, но это еще не значит, что Москва более активна. У Яндекс.Музыки может быть больше пользователей в Москве, поэтому проверю, что величины сопоставимы.

```python
df.groupby('city')['user_id'].nunique()
```

```python
round(df.groupby('city')['genre'].count()/df.groupby('city')['user_id'].nunique(), 2)
```

В среднем за предоставленный период времени 1 пользователь в Москве послушал 1,44 композиции с известным жанром, а в Петербурге 1,48 композиции.

```python
df.groupby('day')['genre'].count()
```

Понедельник и пятница — время для музыки; по средам пользователи немного больше вовлечены в работу. Но нужно проверить это утверждение для каждого города в отдельности.

Создаю функцию *`number_tracks()`*, которая возвращает количество прослушанных композиций в заданный день недели в заданном городе.

```python
def number_tracks(df, day, city):
    return df[(df['day'] == day) & (df['city'] == city)].shape[0]
```

Соберу информацию о количестве прослушанных композиций в таблицу

```python
columns = ['city', 'monday', 'wednesday', 'friday']
data = [['Moscow',
         number_tracks(df, 'Monday', 'Moscow'),
         number_tracks(df, 'Wednesday', 'Moscow'),
         number_tracks(df, 'Friday', 'Moscow')],
        ['Saint-Petersburg',
         number_tracks(df, 'Monday', 'Saint-Petersburg'),
         number_tracks(df, 'Wednesday', 'Saint-Petersburg'),
         number_tracks(df, 'Friday', 'Saint-Petersburg')]]
table = pd.DataFrame(data = data, columns = columns)
table
```

---
**Вывод:**

Результаты показывают, что относительно среды музыку в Петербурге и Москве слушают «зеркально»: в Москве пики приходятся на понедельник и пятницу, а в среду время прослушивания снижается. Тогда как в Санкт-Петербурге среда — день самого большого интереса к музыке, а в понедельник и пятницу он меньше, причём почти одинаково меньше.

## Утро понедельника и вечер пятницы — разная музыка или одна и та же?

Теперь найду ответ на вопрос, какие жанры преобладают в разных городах в понедельник утром и в пятницу вечером. Проверю, отличаются ли жанры в зависимости от города. Кроме того есть предположение, что в понедельник утром пользователи слушают больше бодрящей музыки (например, жанра поп), а вечером пятницы — больше танцевальных (например, электронику).

```python
df.groupby(['city', 'day']).size().unstack(fill_value=0)
```

```python
moscow_df = df[df['city'] == 'Moscow']
```

```python
spb_df = df[df['city'] == 'Saint-Petersburg']
```

Создам функцию *`genre_weekday()`*, которая возвращает список жанров по запрошенному дню недели и времени суток с такого-то часа по такой-то.

```python
def genre_weekday(df, day, time1, time2):
    genre_list = df[(df['day'] == day) & (df['time'] >= time1) & (df['time'] <= time2)]
    genre_list_sorted = genre_list.groupby('genre')['genre'].count().sort_values(ascending = False).head(10)
    return genre_list_sorted
```

Посмотрю, какое самое раннее и самое позднее время есть в данных.

```python
df['time'].min()
```

```python
df['time'].max()
```

Cравню полученные результаты по таблице для Москвы и Санкт-Петербурга в понедельник утром (с 8 до 11) и в пятницу вечером (с 17 до 23).

```python
genre_weekday(moscow_df, 'Monday', '08:00:00', '11:00:00')
```

```python
genre_weekday(spb_df, 'Monday', '08:00:00', '11:00:00')
```

Популярные жанры в понедельник утром в Питере и Москве оказались похожи: везде, как и предполагалось, популярен поп. Несмотря на это, есть и различие: в Питере в топ-10 входит джаз, а в Москве жанр world.

```python
genre_weekday(moscow_df, 'Friday', '17:00:00', '23:00:00')
```

```python
genre_weekday(spb_df, 'Friday', '17:00:00', '23:00:00')
```

В конце недели ситуация не меняется. Поп-музыка всё так же на первом месте. Но в Питере из топ-10 исчезает ruspop и на 10 месте появляется жанр world.

---
**Вывод:**

Гипотеза не подтвердилась. Жанр поп безусловный лидер в обоих городах, а различия в топ-10 минимальны. Однако различия всё же есть: в Москве чаще слушают русскую популярную музыку, в Петербурге — джаз.
